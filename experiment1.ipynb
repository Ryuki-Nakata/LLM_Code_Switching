**実験1:バイリンガルであるというペルソナ設定だけでChatGPTはCode-Switchingを引き起こすのか。**

```python
import json
with open("api_key.json", "r") as f:
    api_key_json = json.load(f)

openai_key = api_key_json["openai"]
```

```python
from openai import OpenAI

client = OpenAI(api_key=openai_key)
```

```python
with open("experiment1_system_japanese_prompt.txt", "r", encoding="utf-8") as f:
    experiment1_system_japanese_prompt = f.read()

with open("experiment1_system_english_prompt.txt", "r", encoding="utf-8") as f:
    experiment1_system_english_prompt = f.read()
```

```python
#質問リストの読み込み
import json
questions_data = json.load(open("questions.json", encoding="utf-8"))

questions_in_english = questions_data["English"]
questions_in_japanese = questions_data["Japanese"]

questions_in_english
```

```python
answers = {}
```

```python
#日本語プロンプト × 日本語質問
for question in questions_in_japanese.values():
    resp = client.chat.completions.create(
        model="gpt-5-mini",
        messages=[
            {"role": "system", "content": experiment1_system_japanese_prompt},
            {"role": "user", "content": question}
        ],
    )
    answers[question] = resp.choices[0].message.content

with open("experiment1_answers_j_j.json", "w", encoding="utf-8") as f:
    json.dump(answers, f, ensure_ascii=False, indent=4)
```
